### Setting the Data so that the all details for one county are in one row

```python
import pandas as pd

# Assuming your original dataset is stored in a DataFrame called 'original_df'

# Create a new DataFrame to store the transformed data
transformed_df = pd.DataFrame()

# Extract the unique county names
county_names = original_df['currently_attending_title'].unique()

# Iterate through each county
for county in county_names:
    county_data = original_df[original_df['currently_attending_title'] == county]
    total_population = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Total', 'currently_attending_Total_astrsk'].values[0]
    male_population = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Male', 'currently_attending_Total_astrsk'].values[0]
    female_population = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Female', 'currently_attending_Total_astrsk'].values[0]
    
    county_info = {
        'County': county,
        'Total_Population': total_population,
        'Male_Population': male_population,
        'Female_Population': female_population
    }
    
    # Iterate through the 26 columns
    for column in ['currently_attending_title', 'currently_attending_County_Sub_County', 'currently_attending_County_Sub_County_original', 'currently_attending_Total_astrsk', 'currently_attending_Pre_Primary', 'currently_attending_Primary', 'currently_attending_Secondary', 'currently_attending_Middle_Level_slash_Technical_and_Vocational_Training_TVET', 'currently_attending_University', 'currently_attending_Adult_Basic_Education', 'currently_attending_Madrasa_slash_Duksi', 'currently_attending_Dont_Know', 'currently_attending_Not_Stated', 'highest_edu_title', 'highest_edu_County_slash_Sub_County', 'highest_edu_Total_astrsk', 'highest_edu_Pre_Primary', 'highest_edu_Primary', 'highest_edu_Secondary', 'highest_edu_Middle Level_Technical_and_Vocational_Training_TVET', 'highest_edu_University', 'highest_edu_Adult_Basic_Education', 'highest_edu_Madrasa_slash_Duksi', 'highest_edu_None', 'highest_edu_Dont_Know', 'highest_edu_Not_Stated']:
        county_info['Total_' + column] = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Total', column].values[0]
        county_info['Male_' + column] = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Male', column].values[0]
        county_info['Female_' + column] = county_data.loc[county_data['currently_attending_County_Sub_County'] == 'Female', column].values[0]

    transformed_df = transformed_df.append(county_info, ignore_index=True)

# Reset the index
transformed_df.reset_index(drop=True, inplace=True)
```
