# Separating the Data

## Group and Conventional Quarters

```python
def get_conventional(x):
    try:
        conventional = x.split(' ')[0]
        return conventional
    except:
        pass

df_edu_house_ownership_minus_total['Conventional']=df_edu_house_ownership_minus_total['Conventional_and_Group_Quarters'].apply(get_conventional)

df_edu_house_ownership_minus_total.head(10)
```

```python
def get_group_quarters(x):
    try:
        group_quarters = x.split(' ')[-1]
        return group_quarters
    except:
        pass

df_edu_house_ownership_minus_total['Group_Quarters']=df_edu_house_ownership_minus_total['Conventional_and_Group_Quarters'].apply(get_group_quarters)

df_edu_house_ownership_minus_total.head(10)
```

## Get Owned, Rented and Not Stated

### Owned

#### Owned Pct

```python
def get_owned_pct(x):
    try:
        owned_pct = x.split(' ')[0]
        return owned_pct
    except:
        pass

df_edu_house_ownership_minus_total['TenureStatus_ConventionalHouseholds_Owned_pct']=df_edu_house_ownership_minus_total['Owned_pct_Rented_num_Rented_pct_Not_Stated_num'].apply(get_owned_pct)

df_edu_house_ownership_minus_total.head(3)
```

### Rented

#### Rented Number

```python
def get_rented_num(x):
    try:
        rented_num = x.split(' ')[1]
        return rented_num
    except:
        pass

df_edu_house_ownership_minus_total['TenureStatus_ConventionalHouseholds_Rented_Slash_Provided_Number']=df_edu_house_ownership_minus_total['Owned_pct_Rented_num_Rented_pct_Not_Stated_num'].apply(get_rented_num)

df_edu_house_ownership_minus_total.head(3)
```

#### Rented Pct

```python
def get_rented_pct(x):
    try:
        rented_pct = x.split(' ')[2]
        return rented_num
    except:
        pass

df_edu_house_ownership_minus_total['TenureStatus_ConventionalHouseholds_Rented_Slash_Provided_Pct']=df_edu_house_ownership_minus_total['Owned_pct_Rented_num_Rented_pct_Not_Stated_num'].apply(get_rented_pct)

df_edu_house_ownership_minus_total.head(3)
```

### Not Stated

#### Not Stated Number

```python
def get_not_stated_num(x):
    try:
        not_stated_num = x.split(' ')[2]
        return not_stated_num
    except:
        pass

df_edu_house_ownership_minus_total['TenureStatus_ConventionalHouseholds_Rented_Slash_Provided_Not_Stated_Number']=df_edu_house_ownership_minus_total['Owned_pct_Rented_num_Rented_pct_Not_Stated_num'].apply(get_not_stated_num)

df_edu_house_ownership_minus_total.head(3)
```
