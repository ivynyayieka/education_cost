# Reading the Data

```python
df_edu_lighting_fuel=pd.read_csv("edu_lighting_fuel.csv")
df_edu_piped=pd.read_csv("edu_piped.csv")
df_edu_internet=pd.read_csv("edu_internet.csv")
df_edu_mobile_phone=pd.read_csv("edu_mobile_phone.csv")
```

# Merging the data

```python
import pandas as pd

# Assuming you have the four datasets: df_edu_lighting_fuel, df_edu_piped, df_edu_internet, df_edu_mobile_phone

# Merge the datasets one by one using the 'County_slash_Sub_County' column
merged_df = df_edu_lighting_fuel.merge(df_edu_piped, on='County_slash_Sub_County', how='inner')
merged_df = merged_df.merge(df_edu_internet, on='County_slash_Sub_County', how='inner')
merged_df = merged_df.merge(df_edu_mobile_phone, on='County_slash_Sub_County', how='inner')

# Exclude columns where 'County_slash_Sub_County' value contains 'KENYA FOREST'
merged_df = merged_df.loc[~merged_df['County_slash_Sub_County'].str.contains('KENYA FOREST')]

# Print the resulting merged DataFrame
merged_df
```

```python
merged_df.to_csv("all_counties.csv")
```
