```python
df_school_unclean=pd.read_csv("untitled_school_attendance_age.csv")
```

```python
def get_title(x):
    try:
        title = x.split(' ')[0]
        return title
    except:
        pass

df_school_unclean['title']=df_school_unclean['County/ Special Age Groups  Total*'].apply(get_title)

df_school_unclean.head(10)
```
