# Change the column titles

```python
import pandas as pd

def rename_columns(original_df):
    # Create a mapping dictionary for column name replacements
    mapping = {col: col.replace(' ', '_').replace('/', '_').replace('*', '_astrsk') for col in original_df.columns}
    
    # Rename the columns using the mapping dictionary
    original_df.rename(columns=mapping, inplace=True)

# Example usage
# Assuming you have a DataFrame named 'df'
rename_columns(original_df)
```

### Setting the Data so that the all details for one county are in one row

```python
import pandas as pd

# Assuming your original dataset is stored in a DataFrame called 'original_df'
# and the age group details are stored in a column called 'title_and_age'
# and the county titles are stored in a column called 'title'

# Define the age groups
age_groups = {
    'Total': 'Total',
    '3-5': '3to5',
    '4-5': '4to5',
    '6-13': '6to13',
    '14-17': '14to17',
    '18-22': '18to22',
    '23+': '23plus',
    'Not Stated': 'NotStated'
}

# Create an empty dictionary to store the transformed data
transformed_data = {}

# Extract the unique county names
county_names = original_df['title'].unique()

# Initialize the transformed_data dictionary with default values for each column
for county in county_names:
    transformed_data[county] = {'County': county}
    for age_group, prefix in age_groups.items():
        for column in original_df.columns:
            transformed_data[county][prefix + '_' + column] = []

# Iterate through each row in the original dataframe
for index, row in original_df.iterrows():
    county = row['title']
    age_group = row['title_and_age']

    # Iterate through each column in the row
    for column in original_df.columns:
        transformed_data[county][age_groups[age_group] + '_' + column].append(row[column])

# Create the transformed DataFrame
transformed_df = pd.DataFrame()
for county, data in transformed_data.items():
    county_df = pd.DataFrame(data)
    transformed_df = transformed_df.append(county_df, ignore_index=True)

# Reset the index
transformed_df.reset_index(drop=True, inplace=True)
```
